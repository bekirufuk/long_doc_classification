```python
import pandas as pd
import numpy as np
import pickle
from scipy.sparse import csr_matrix
```

```python
partition = 'test'
```

```python
tfidf = pd.DataFrame()
for i in range(8):
    print(i)
    df = pd.read_pickle('../data/refined_patents/tfidf/longformer_tokenizer_no_stopwords/label_based/'+partition+'_'+str(i)+'_tfidf.pkl')  
    print('concat')
    tfidf = pd.concat([tfidf, df], ignore_index=True)
```

```python
tfidf.set_index('general_index')
```

```python
tfidf = tfidf.sort_index()
#tfidf = tfidf.fillna(0)
tfidf.index.name = None
```
